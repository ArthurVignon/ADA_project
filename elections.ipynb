```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from pandas.io.json import json_normalize
```

```python
PATH_ELECTIONS_2015 = "data/elections/elections_2015.csv"
PATH_ELECTIONS_2019 = "data/elections/elections_2019.csv"

romand_cantons = ['Vaud','Jura','Genève','Valais / Wallis','Neuchâtel','Fribourg / Freiburg']
selected_parties = ['SVP','SP','FDP','CVP','GPS','GLP']
columns = ['wahl_jahr','kanton_bezeichnung','partei_id','partei_staerke', 
           'partei_bezeichnung_fr','partei_bezeichnung_en','fiktive_waehlende']
```

```python
elections_2015 = pd.read_csv(PATH_ELECTIONS_2015, sep=';')
elections_2015
```

```python
elections_2015.columns
```

```python
elections_2015 = elections_2015[columns]
elections_2015
```

```python
elections_2019 = pd.read_csv(PATH_ELECTIONS_2019, sep=';')
elections_2019
```

```python
elections_2019.columns
```

```python
elections_2019 = elections_2019[columns]
elections_2019
```

```python
elections = pd.concat([elections_2015, elections_2019])
elections = elections.loc[elections['partei_bezeichnung_en'].isin(selected_parties)]
elections = elections.sort_values(by='partei_staerke',ascending=False)
elections
```

```python
swiss_elections = elections[elections['kanton_bezeichnung'] == 'Schweiz']
swiss_elections
```

```python
swiss_elections_2015 = swiss_elections.loc[swiss_elections['wahl_jahr'] == 2015]
swiss_elections_2015
```

```python
swiss_elections_2019 = swiss_elections.loc[swiss_elections['wahl_jahr'] == 2019]
swiss_elections_2019
```
